```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor as RF
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('train.csv')
data.describe()
data
```

```python
data.columns
```

```python
features = ['Id','MSSubClass','1stFlrSF','2ndFlrSF','FullBath','MoSold','YrSold']
X = data[features]
X.head()
```

```python
y = data.SalePrice
y.head()
```

```python
train_X,val_X,train_y,val_y = train_test_split(X,y)
```

```python
def get_mae(num_of_leaves):
      model = RF(max_leaf_nodes=num_of_leaves,random_state=1)
      model.fit(train_X,train_y)
      preds = model.predict(val_X)
      mae = mean_absolute_error(val_y,preds)
      return mae
```

```python
get_mae(50)
```

```python
max_no_nodes =[100,150,200,250,300,350,400,450,500]
for  i in max_no_nodes:
      min = 999999
      id = 0
      err = get_mae(i)
      if err<min:
            min = err
            id = i
id
```

```python
full_model = RF(max_leaf_nodes=id,random_state=1)
full_model.fit(X,y)
```

```python
test_Data = pd.read_csv('test.csv')
```

```python
test_Data
```

```python
ans = full_model.predict(test_Data[features])
ans
```

```python
import csv
with open('submission.csv','w') as f:
      writer = csv.writer(f)
      writer.writerow(['Id','SalePrice'])
      for i,j in zip(test_Data.Id, ans):
            writer.writerow([i,j])
```
